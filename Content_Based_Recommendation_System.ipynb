```python
movies['keywords'] = movies['overview']+movies['genre']
```

```python
new_data=movies.drop(columns=['overview','genre','original_language','popularity','release_date','vote_average','vote_count'])
```

```python
new_data.head(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000,stop_words='english')
```

```python
vector=cv.fit_transform(new_data['keywords'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['title']=="The Godfather"]
```

```python
distance = sorted(list(enumerate(similarity[2])),reverse=True,key=lambda vector:vector[1])
for i in distance[0:5]:
  print(new_data.iloc[i[0]].title)
```

```python
def recommend(movies):
  index=new_data[new_data['title']==movies].index[0]
  distance = sorted(list(enumerate(similarity[index])), reverse-True, key=lambda vector:vector[1])
  for i in distance[1:6]:
    print(new_data.iloc[i[0]].title)

    #Get user input
    movie_name = input("Enter a movie title ")

    recommend(movie_name)
```

```python
def recommend(movies):
  movies = movies.lower()
  new_data['title'] = new_data['title'].str.lower()
  index = new_data[new_data['title'] == movies].index[0]
  distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
  for i in distance[1:6]:
    print(new_data.iloc[i[0]].title)

movie_name = input("Enter a movie title ")
recommend(movie_name)
```
